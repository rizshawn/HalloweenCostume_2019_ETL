```python
import pymongo
from splinter import Browser
from bs4 import BeautifulSoup
import os
from datetime import datetime
import platform
import pandas as pd


def init_browser():
    if platform.system().lower() == 'windows'.lower():
        executable_path = {
            'executable_path': 
            os.path.join(os.getcwd(), 'chromedriver.exe')}
        return Browser('chrome', **executable_path, headless=False)
    else:
        return Browser('chrome')
    
def get_html(browser, url):
    browser.visit(url)
    html = browser.html
    return html

def get_listings(html):
    soup = BeautifulSoup(html, "html.parser")
    listings = soup.find_all(
        'a', class_='thumb-link')
    return listings

def scrape(browser, url):
    html = get_html(browser, url)
    listings = get_listings(html)
    return listings

def access_db(dbname, collectionname):
    conn = 'mongodb://localhost:27017'
    client = pymongo.MongoClient(conn)
    db = client[dbname]
    collection = db[collectionname]
    return db, collection

def scrape_many(search_terms, collection, browser, base_url, last_url):
    for search_term in search_terms:
        url = base_url + search_term + last_url
        titles = scrape(browser, url)
        time = datetime.now()
        collection.insert_many([
            {'searchterm':search_term, 'title':title.text, 
             'datetime':time, 'order':order
            } for order, title in enumerate(titles)])
        
def main():
    db, collection = access_db('partycity', 'searchrank')
    browser = init_browser()
    base_url = "https://www.partycity.com/"
    last_url = "-costumes?srule=top-rated"
    search_terms = ['womens', 'mens', 'kids', 'couples']
    scrape_many(search_terms, collection, browser, base_url,last_url)
    return 
main()

db, collection = access_db('partycity','searchrank')
[data['searchterm'] for data in collection.find()]

data = list(collection.find())

party_city = pd.DataFrame(data)
```
